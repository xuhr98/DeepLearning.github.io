```python
#训练一个模型的流程
#1 prepare dataset
#2 design model using class
#3 construct loss and optimizer(using pytorch api)
#4 training circle

#1 prepare dataset
import torch
x_data=torch.rand(3,1,requires_grad=True)
y_data=torch.rand(3,1,requires_grad=True)
```

```python
#2 design model using class
class LinearModel(torch.nn.Module):
    
    def __init__(self):   #init前后2个下划线
        super(LinearModel,self).__init__()
        self.linear=torch.nn.Linear(1,1)
    
    def forward(self,x):
        y_pred=self.linear(x)
        return y_pred
    
model=LinearModel()
```

```python
#3 construct loss and optimizer(using pytorch api
criterion = torch.nn.MSELoss(reduction='sum') #求（y-y_pred）^2

optimizer = torch.optim.SGD(model.parameters(), lr=0.01)
```

```python
#4 Training Cycle

for epoch in range(100):
    y_pred=model(x_data)
    loss= criterion(y_pred,y_data)
    print(epoch,loss.item())
    
    optimizer.zero_grad() #清0梯度
    loss.backward()
    optimizer.step()
    
    
    
```
