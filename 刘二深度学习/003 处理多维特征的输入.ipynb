```python
#1 prepare dataset
import numpy as np
import pandas as pd
import torch
#x1=np.loadtxt('X.csv', delimiter=',', dtype=str)
#y1=np.loadtxt('y.csv', delimiter=',', dtype=str)

x1=pd.read_csv('X.csv')
y1=pd.read_csv('y.csv')

def custom_converter(value):
    try:
        return float(value)
    except ValueError:
        return None
    
x1 = x1.applymap(custom_converter)
numpy_array = x1.values.astype(np.float32)
x_data=torch.tensor(numpy_array,dtype=torch.float32) #最后一列不要，因为是y

y1 = y1.applymap(custom_converter)
numpy_array = y1.values.astype(np.float32)
y_data=torch.from_numpy(numpy_array) #【0】代表y取成矩阵

print(x_data)
```

```python
#2 define model

class Model(torch.nn.Module):
    def __init__(self):
        super(Model,self).__init__()
        self.linear1=torch.nn.Linear(441,6)
        self.linear2=torch.nn.Linear(6,4)
        self.linear3=torch.nn.Linear(4,1)
        self.sigmoid=torch.nn.Sigmoid()
        
    def forward(self,x):
        x=self.sigmoid(self.linear1(x))
        x=self.sigmoid(self.linear2(x))
        x=self.sigmoid(self.linear3(x))  #其实这里是x1,x2,y_pred但都写成x是惯例，防止打字错误
        return x
    
model=Model()
```

```python
criterion=torch.nn.BCELoss(reduction='sum') #求交叉熵
optimizer=torch.optim.SGD(model.parameters(),lr=0.01)
```

```python
for epoch in range(100):
    y_pred=model(x_data)
    loss= critetion(y_pred,y_data)
    print('epoch=%d,loss=%f' %(epoch,loss.item()))
    
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
```
